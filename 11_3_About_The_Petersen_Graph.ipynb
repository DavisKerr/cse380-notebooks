```
petersen_graph_nodes = [('a', "3,4!"), ('b', "7,3!"), ('c', "6,0!"), ('d', "1,0!"), ('e', "0,3!"), ('f', "3,3!"), ('g', "5,2!"), ('h', "4,1!"), ('i', "2,1!"), ('j', "1,2!")]
petersen_graph_edges = ['ab', 'af', 'ae', 'bg', 'bc', 'ch', 'cd', 'di', 'de', 'ej', 'fh', 'fi', 'gj', 'gi', 'hj']
dot = Graph(format='png', engine="neato")
for node in petersen_graph_nodes:
  dot.node(node[0], pos=node[1])
dot.edges(petersen_graph_edges)
dot.render('test-output/graph', view=True)
display(Image('/content/test-output/graph.png'))
```
