### Is It Decidable?

#### TODO Study and run the code below.

And answer these questions:
1. What is this program computing?
2. How does it do it?
3. Will it ever terminate?

```python
from math import log2

def you_decide():
  flibbers = [17,78,19,23,29,77,95,77,1,11,13,15,15,55]
  flobbers = [91,85,51,38,33,29,23,19,17,13,11,14,2,1]
  i = 2
  time_keeps_ticking = True
  while time_keeps_ticking:
    time_keeps_ticking = False
    for n in range(len(flibbers)):
      f = i * flibbers[n]
      if not f % flobbers[n]:
        i = f // flobbers[n]
        p = log2(i)
        if p.is_integer():
          print(i)
        time_keeps_ticking = True
        break
```

```python
you_decide()
```

1. The program outputs the prime powers of 2 in ascending order starting with the prime of 2. For example, $2^2, 2^3, 2^5, 2^7, 2^{11}, etc. $

2. It makes use of Conway's fractran, a generalization of the Collatz function. The number f is a number i * any number in fibbers. The number i updates only if f is divisable by any number in flobbers. The last digit of flobbers is 1, so f will always be divisable by at least 1 element in flobbers. This will mean the number p (log2 of i) is always prime according to my understanding of Conway's Fractran. If the log2 of i is a whole number, i is a prime power of 2. 

3. The program will never terminate becasue the last digit to mod our number by is 1, so time keeps ticking will always be true.
