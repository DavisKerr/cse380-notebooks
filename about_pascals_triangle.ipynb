## A Helper Function
Exercise 253 on page 161 shows a recursive definition of `nCk`. Here's a much more efficient way to compute binomial coefficients, which perhaps you can profitably use in your investigations:


```python
from math import gcd

def nCk(n, k):
  if k < 0 or k > n:
    return 0
  else:
    result = 1
    d = 1
    g = 1
    m = min(k, n - k)
    while (d <= m):
      g = gcd(result, d)
      result = n * (result // g)
      result = (result // (d // g))
      n -= 1
      d += 1
    return result
```

```python
def check_row_for_fib(row):
  if len(row) > 2:
    start = 0
    end  = 2
    while end < len(row):
      if row[start] + row[start + 1] == row[end]:
        return True
      else:
        end += 1
        start += 1
    return False

def check_row_for_ratio(row):
  if len(row) > 2:
    start = 0
    end  = 2
    while end < len(row):
      if (row[start] / row[start + 1]) == (1/2) and (row[start + 1] / row[end]) == 2/3:
        return True
      else:
        end += 1
        start += 1
    return False

def gen_tri(rows):
  n = 0
  while n < rows:
    row = []
    for k in range(0, n + 1):
      row.append(nCk(n,k))
    if check_row_for_fib(row):
      print(f'At row {n} there are fibanacci numbers:')
      print(f'\t{row}')
    if check_row_for_ratio(row):
      print(f'At row {n} there are 1:2:3 ratio numbers:')
      print(f'\t{row}')
      #print(f'{nCk(n,k)}', end=" ")
    n += 1
    #print(f'\n')


gen_tri(900)
```

## Predominant and Fundamental
The predominant pattern, the fundamental symmetry in Pascal's triangle is captured by the identity:
$${n \choose k} = {n \choose n - k}$$
## Identify Identities
Here are two different identities that point out the same pattern in Pascal's triangle. What is the name of this pattern?
### Identity 1
For nonnegative integers $r$ and $n$:
$$\sum^n_{k=0} {r+k \choose k} = {r+n+1 \choose n}$$
### Identity 2
For nonnegative integers $m$ and $n$:
$$\sum^n_{k=0} {k \choose m} = {n+1 \choose m+1}$$
